#Exercício - Bibliotecas NumPy e Matplotlib

Imprima um gráfico de três funções exponenciais: exp(x), exp(2x) e exp(3x), para x de 0 a 1. Mude os limites de impressão dos eixos para que todos os gráficos apareçam na figura com boa visualização de suas diferenças. Use valores distintos para as espessuras da linha de impressão das funções (espessuras de 1 a 3), e crie legenda.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)
f1 = np.exp(x)
f2 = np.exp(2*x)
f3 = np.exp(3*x)

plt.subplot()
plt.plot(x, f1, label='Função $e^x$', linewidth = 1.0)
plt.plot(x, f2, label='Função $e^{2x}$', linewidth = 2.0)
plt.plot(x, f3, label='Função $e^{3x}$', linewidth = 3.0)
plt.title('Exercício - Biblioteca Numpy e Matplotlib')
plt.xlabel('x')
plt.ylabel('Função')
plt.axis([0, 1, 0, 10]) #Coloca o limite nos eixos
plt.legend()
plt.grid()
plt.show()
```

#Outros Exemplos do Matplotlib

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
f1 = x**2
f2 = x**3

fig, ax = plt.subplots(1, 2)
ax[0].plot(x, f1)
ax[1].plot(x, f2)
ax[0].set_title('f(x)=$x^2$')
ax[1].set_title('f(x)=$x^3$')
ax[0].set_ylabel('Aceleração [m/$s^2$]')
ax[1].set_ylabel('Aceleração [m/$s^2$]')
plt.tight_layout(pad = 1.0) #Para nome do eixo não sobrepor o outro gráfico, esse comando resolve isso
plt.show()
```

#Exercício - Imprimindo-se um gráfico com ruído

Utilize a biblioteca Matplolib para imprimir o gráfico das funções cúbica e quadrática de "x" em uma só figura, um em cada sub-figura. Utilize a biblioteca "Random" um laço "for" para adicionar um suave ruído sobreposto à esses gráficos. Adicione título e rótulos aos eixos de cada gráfico, separadamente, e um título principal na região superior da figura.

OBS1: Utilize um número de pontos "x" (que é o argumento da função) que permita a visualização do ruído (não use muitos pontos).
OBS2: Você pode criar números aleatórios inteiros, e multiplica-los por um vqalor menor que 1 para adicionar um ruído suave.

```python
import matplotlib.pyplot as pl
import numpy as np
import random

x = np.linspace(0, 1, 100)
f1 = x**2 + random.randint(0, 5)
f2 = x**3 + random.randint(0, 5)


fig, ax = plt.subplots(1, 2)
ax[0].plot(x, f1)
ax[1].plot(x, f2)
ax[0].set_title('f(x)=$x^2$')
ax[1].set_title('f(x)=$x^3$')
ax[0].set_ylabel('Aceleração [m/$s^2$]')
ax[1].set_ylabel('Aceleração [m/$s^2$]')
plt.tight_layout(pad = 1.0) #Para nome do eixo não sobrepor o outro gráfico, esse comando resolve isso
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando valores de x
x = np.linspace(0, 10, 100)

# Definindo as funções f1 e f2
f1 = x**2
f2 = x**3

# Gerando ruído
import random as rd
ruído_f1 = [rd.randint(-10, 10) for _ in range(100)]
ruído_f2 = [rd.randint(-10, 10) for _ in range(100)]

# Adicionando ruído às funções
f1_com_ruído = f1 + ruído_f1
f2_com_ruído = f2 + ruído_f2

# Criando os gráficos
fig, ax = plt.subplots(1, 2)
ax[0].plot(x, f1_com_ruído)
ax[1].plot(x, f2_com_ruído)
ax[0].set_title('f(x)=$x^2$ com Ruído')
ax[1].set_title('f(x)=$x^3$ com Ruído')
ax[0].set_ylabel('Aceleração [m/$s^2$]')
ax[1].set_ylabel('Aceleração [m/$s^2$]')
plt.tight_layout(pad=1.0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random as rd

x = np.linspace(0, 1, 100)
f1 = x**2
f2 = x**3

for n in range(len(f1)):
  f1[n] += 0.01*rd.randint(-10, 10)
  f2[n] += 0.01*rd.randint(-10, 10)

plt.plot(x, f1, x, f2)
plt.show()
```
