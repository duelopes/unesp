#Matplotlib - Imprimindo-se em Sub-Figuras

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 4, 100)

fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(x, np.exp(-x))
ax[0, 1].plot(x, np.exp(-2*x))
ax[1, 0].plot(x, np.exp(-3*x))
ax[1, 1].plot(x, np.exp(-4*x))

ax[0, 0].set_title('$e^{-x}$')
ax[0, 1].set_title('$e^{-2x}$')
ax[1, 0].set_title('$e^{-3x}$')
ax[1, 1].set_title('$e^{-4x}$')
plt.tight_layout(pad = 1.5) #Para nome do eixo não sobrepor o outro gráfico, esse comando resolve isso
fig.suptitle('Título Superior')
plt.show()
```

#Exercício - Imprimindo-se em Sub-Figuras

Desenvolva um algoritmo que imprima senos em Sub-Figuras, com um número de linhas e colunas variável (pode se mudar a variável do número de linhas e colunas da Figura) alterando-se uma variável numérica (por exemplo, linha=2 e colunas=3). Faça com o argumento do seno aumente de pi/2 a cada coluna que se aumente (começando-se em 0), e o argumento do seno também seja multiplicado pelo número de linha impressa (começando-se em 1).

##Primeira tentativa

Testando o cálculo do exercício, usando o arquivo da aula 8:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
fig, ax = plt.subplots(3, 3)

for linha in range(3):
    for coluna in range(3):
      ax[linha, coluna].plot(x, np.sin(x+(np.pi/2*coluna))*linha)
      plt.tight_layout(pad = 1.0)


fig.suptitle('Resultado do Exercício')
plt.tight_layout(pad=1.5)
plt.show()
```

##Segunda tentativa

Ajustando a imagem:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)

fig, ax = plt.subplots(3, 3, figsize=(10, 8))

for linha in range(3):
    for coluna in range(3):
        ax[linha, coluna].plot(x, np.sin(x+((np.pi/2)*coluna))*linha)
        ax[linha, coluna].set_title(f'Linha {linha}, Coluna {coluna}')

fig.suptitle('Resultado do Exercício', fontsize=16)
plt.tight_layout(pad=1.5)
plt.show()
```

##Terceira Tentativa

Refinando o código para receber do usuário o número de colunas e de linhas, utilizando o arquivo da aula 2:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = int(input('Número de linhas e colunas: '))

fig, ax = plt.subplots(y, y, figsize=(10, 8))

for linha in range(y):
    for coluna in range(y):
        ax[linha, coluna].plot(x, np.sin(x+((np.pi/2)*coluna))*(linha+1))
        ax[linha, coluna].set_title(f'Linha {linha+1}, Coluna {coluna}')

fig.suptitle('Resultado do Exercício', fontsize=16)
plt.tight_layout(pad=1.5)
plt.show()
```

##Resultado Final

Refinando o código para receber do usuário o número de linhas e colunas:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
n_linhas = int(input('Número de linhas: '))
n_colunas = int(input('Número de colunas: '))

fig, ax = plt.subplots(n_linhas, n_colunas, figsize=(10, 8))

for linha in range(n_linhas):
    for coluna in range(n_colunas):
        ax[linha, coluna].plot(x, np.sin(x+((np.pi/2)*coluna))*(linha+1))
        ax[linha, coluna].set_title(f'sen(x+{coluna}*(π/2))*{linha+1}')

fig.suptitle('Resultado do Exercício', fontsize=16)
plt.tight_layout(pad=1.5)
plt.show()
```

#Exercício - De Volta ao "For" (Desconsiderar)

Desenvolva um algoritmo que imprima três gráfico em uma só linha de uma figura, ponto a ponto, utilizando-se somente um "for". O primeiro gráfico será os valores numéricos do iterável "n" do "for", o segundo gráfico será o argumento "x" de uma função (de 0 a 1, com 11 valores), e o terceiro gráfico será a função "f(x) = x ao quadrado".

OBS1: Para se imprimir a função, defina ela antes, no início do programa, e vá chamando-a ao longo do "for".
OBS2: Utilize o iterável como "eixo x" de cada um dos gráficos, e o mais importante, interprete os valores do gráfico, e escreva uma breve discussão sobre a diferença entre "x", o valor do iterável e "f(x)" nesse caso, e sobre a diferença entre "f[n]" e "f[x]"

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 11)
fig, ax = plt.subplots(1, 3)

for n in range(10):
  ax[0].plot(n, n, marker='s', markersize=4.0)
  ax[1].plot(x, n, marker='s', markersize=4.0)
  ax[2].plot(x, x**2, marker='s', markersize=4.0)

fig.suptitle('Resultado do Exercício', fontsize=16)
plt.tight_layout(pad=1.5)
plt.show()
```
