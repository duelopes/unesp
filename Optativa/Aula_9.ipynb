## Testando o Matplolib

Através da página do Matplolib conseguimos modelos para testar

## Exercício - Biblioteca NumPy e Matplotlib

Utilize as bibliotecas do Numpy e do Matplotlib para se imprimir um gráfico de três senos, cada um espaçado de pi/2 radianos do anterior. Adicione um título para o gráfico, um rótulo para o eixo 'x' um rótulo para o eixo 'y', e use a documentação e os exemplos da biblioteca para se obter uma legenda de 3 funções impressas.

**OBS:** verifique com quantos pontos o gráfico parece continuo, quando não é mais possível se distinguir os valores discretos

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 50)
f1 = np.sin(x)
f2 = np.sin(x-(np.pi/2))
f3 = np.sin(x-(np.pi/4))

plt.subplot()
plt.plot(x, f1, x, f2, x, f3)
plt.title('Exercício - Biblioteca Numpy e Matplolib')
plt.xlabel('x')
plt.ylabel('Função')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 50)
f1 = np.sin(x)
f2 = np.sin(x-(np.pi/2))
f3 = np.sin(x-(np.pi/4))

plt.subplot()
plt.plot(x, f1, label='sen(x)')
plt.plot(x, f2, label='sen(x-π/2)')
plt.plot(x, f3, label='sen(x-π/4)')
plt.title('Exercício - Biblioteca Numpy e Matplotlib')
plt.xlabel('x')
plt.ylabel('Função')
plt.legend()  # Adiciona a legenda
plt.grid()
plt.show()
```

Utilizando a biblioteca para deixar mais atraente

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 50)
f1 = np.sin(x)
f2 = np.sin(x-(np.pi/2))
f3 = np.sin(x-np.pi)

plt.subplot()
plt.plot(x, f1, label='sen(x)', linestyle='-', color='blue')
plt.plot(x, f2, label='sen(x-π/2)', linestyle='--', color='green')
plt.plot(x, f3, label='sen(x-π)', linestyle=':', color='red')
plt.title('Exercício - Biblioteca Numpy e Matplotlib')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right', fontsize='small')
plt.grid()
plt.tight_layout()
plt.gca().set_facecolor('lightgray')
plt.show()
```
